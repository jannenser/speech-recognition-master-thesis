```python
from tqdm import tqdm
names = ds["train"]["text_swiss"]

# open file in write mode
with open(r'texts.txt', 'w') as fp:
    for item in tqdm(names, len(names)):
        fp.write("%s\n" % item)
    print('Done')
```

```python
import nltk


with open(r'output.txt', 'w') as wf:
    with open(r'/kaggle/input/d/andreymesser/swiss-text/texts.txt', 'r') as rf:
        for line in rf:
            for sentence in nltk.sent_tokenize(line):
                item = str(' '.join(nltk.word_tokenize(sentence)).lower())
                wf.write("%s\n" % item)
```
